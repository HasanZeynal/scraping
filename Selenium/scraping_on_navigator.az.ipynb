```python
#path = '/usr/local/bin/chromedriver'
driver = webdriver.Chrome(executable_path= ChromeDriverManager().install())
path = 'https://navigator.az/az/scat/14/avtoservis-ve-diaqnostika/'

#creating lists
all_link_urls = []
names = []
tags = []
phone_numbers = []
subways = []
locations = []

#iteration on all pages
for i in range(1,10):
    driver.get(path + str(i))
    time.sleep(2)
    links = driver.find_elements_by_class_name('main-card__title')
    link_urls = [link.get_attribute('href') for link in links]
    
    #iteration on all classes by every pages
    for i in link_urls:
        all_link_urls.append(i)
        driver.get(i)
        time.sleep(2)
        #getting the names
        name = driver.find_element_by_class_name('company-name')
        names.append(name.text)
        #getting the tags
        tag = driver.find_element_by_class_name('tags')
        tags.append(tag.text)
        #getting locations
        location = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div/section[5]/div/div/div/div[1]/div[2]/p')
        locations.append(location.text)
        #getting subways(there is no subway info in some of companies so we have to catch error)
        try:
            subway = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div/section[5]/div/div/div/div[4]/div[2]/p')
            subways.append(subway.text)
        except NoSuchElementException:  
            subways.append('Not Info')

            
        phone_number = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div/section[5]/div/div/div/div[3]/div[2]/p')
        phone_numbers.append(phone_number.text)
driver.quit()
```

```python
#creating dataframe
df = pd.DataFrame(data = (list(zip(all_link_urls,names,tags,phone_numbers,subways,locations))),
            columns=['link_url','company_name','tags','phone_numbers','subway','location'])
df.head()
```

```python
#replace /n 
df['tags'] = df['tags'].str.replace('\n',',')
df['phone_numbers'] = df['phone_numbers'].str.replace('\n',',')
df['location'] = df['location'].str.replace('\n','')
```
