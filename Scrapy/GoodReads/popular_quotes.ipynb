```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv('goodreads.csv')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
list1 = []
for i in df['author name']:
    list1.append(i)
list1
```

```python
import re
re.findall(r'[\n]','\n    H. Jackson Brown Jr.,\n  ')
```

```python
df['author name'] = df['author name'].str.replace("[\n]","")
```

```python
df['author name'] = df['author name'].str.replace("  ","")
```

```python
df['sentence'] = df['sentence'].str.replace("[\n]","")
```

```python
df['sentence'] = df['sentence'].str.replace("―","")
df['sentence'] = df['sentence'].str.replace("  ","")
```

```python
df['likes'] = df['likes'].str.replace('likes','')
```

```python
df.columns
```

```python
df.rename({'sentence':'quote',
          'author name':'author_name',
           'likes':'number_of_likes',
          'image link':'image_link'},inplace=True,axis=1)
```
